```python
numerator = sum((x[i] - mean_x) * (y[i]-mean_y) for i in range(n))
denominator = sum((x[i]-mean_x)**2 for i in range(n))
```

```python
m = numerator/denominator
```

```python
m
```

```python
c = mean_y - m * mean_x
```

```python
c
```

```python
print(f'Regression Line y = {m}*x+{c}')
```

```python
predicted_marks = [m*i+c for i in x]
```

```python
predicted_marks
```

```python
for i in range(n):
    print(f"Actual marks {y[i]} and Predicted Marks {predicted_marks[i]}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,color='blue',label='Actual')
plt.plot(x,predicted_marks,color='red',label='Regression Line')         
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
cleaned_df = df.drop('Address',axis=1)
```

```python
cleaned_df
```

```python
cleaned_df.corr().style.background_gradient()
```

```python
X = cleaned_df.iloc[:,:-1]
```

```python
y = cleaned_df['Price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2)
```

```python
# len(xtrain)

xtrain
```

```python
len(ytest)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression() #object creation
```

```python
lr.fit(xtrain,ytrain) #training the model on training data
```

```python
predict = lr.predict(xtest) # doing prediction on test data
```

```python
len(predict)
```

```python
predict
```

```python
plt.scatter(range(len(ytest)),ytest,color='blue',label='Actual Price')
plt.scatter(range(len(predict)),predict,color='orange',label='Predicted Price')
plt.grid()
plt.legend()
plt.show()
```

1. MAE  (Mean Absolute Error)

2. MSE  (Mean Squared Error)

3. RMSE (Root Mean Squared Error)

4. R2 Score 

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error(ytest,predict)
mse = mean_squared_error(ytest,predict)
rmse = np.sqrt(mse)
r2 = r2_score(ytest,predict)
```

```python
print(f"Mean Absolute Error {mae}")
print(f"Mean Squared Error {mse}")
print(f"Root Mean Squared Error {rmse}")
print(f"R2 Score {r2}")
```
